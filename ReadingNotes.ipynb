# The Garbage Collection Handbook
## 术语
1. 堆是由一段或者几段连续内存组成的空间集合。
2. 对象是为应用程序分配的内存单元，常常为可寻址的**连续字节或字的数组**，内部划分为多个**槽, 域**。
3. 内存块是依照特定大小对齐的大块内存。
4. 堆通常可以使用对象图（有向图）描述，有向边描述对象之间的引用。
5. 赋值器执行应用代码，分配对象，修改对象图的拓扑结构。
6. 回收器执行垃圾回收代码。**一个程序中可以由多个赋值器和多个回收器组成，但是仅可以有一个堆。**
7. 分配器执行2种操作：分配(allocate)、释放(free)
8. 赋值器包括3种与回收器相关的操作：创建(New)、读(Read)、写(Write)


## 标记—清扫回收
标记-清扫算法\[McCarthy, 1960\]是一种间接回收算法，通过确定所有存活对象，反过来判定其余对象都是垃圾的算法。
### 回收过程
回收过程分为两个阶段，第一阶段为追踪。回收器从根集合开始遍历对象图并且对每一个可以到达的对象打上标记。第二阶段为清扫，回收器检查堆中每一个对象并且把所有未标记的对象当作垃圾回收。

### 分配New()步骤
1. 申请堆空间
2. 如果申请成功则执行6
3. 调用回收器
4. 重新申请堆空间
5. 如果申请失败则抛出异常（堆空间满）
6. 返回申请结果

### 回收器Collect()步骤
**整个过程原子化**  
1. 从根节点开始遍历可以访问的对象并且打上标记
2. 从堆头到堆尾清理未被标记的对象
